```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
#file_to_load = "Resources/purchase_data.csv"
file_to_load="purchase_data.csv"

# Read Purchasing File and store into Pandas data frame
purchase_df = pd.read_csv(file_to_load)
```

```python
#getting the unique number of players using screen name
total_players=purchase_df['SN'].nunique()
total_players
```

```python
#making a dataframe for player count
#player count table
total_players2=[{'Total Players':total_players}]
TP_df=pd.DataFrame(total_players2)
TP_df
```

```python
#finding the unique item id
item_id=purchase_df['Item ID'].nunique() 
item_id
```

```python
#finding the average price
avg_price=purchase_df['Price'].mean() 
avg_price
```

```python
#getting the total revenue
total_revenue=purchase_df['Price'].sum() 
total_revenue
```

```python
#getting the count of number of purchases
no_of_purchases=purchase_df['Purchase ID'].count()
no_of_purchases
```

```python
#creating a dataframe to get the table of purchasing analysis
#purchasing analysis df
table2=[{'Number of Unique Items':item_id, 'Average Price':avg_price, 'Number of Purhases':no_of_purchases,
        'Total Revenue':total_revenue}]
table2_new_df=pd.DataFrame(table2)
#formatting
table2_new_df["Average Price"] = table2_new_df["Average Price"].map("${:.2f}".format)
table2_new_df["Total Revenue"] = table2_new_df["Total Revenue"].map("${:.2f}".format)
table2_new_df
```

```python
#purchasing Analysis(Gender)
purchase_count=purchase_df['Gender'].value_counts()
purchase_count
```

```python
#grouping by gender and getting price for each category in gender
#new_purchase_count=purchase_df.groupby(['Gender'])
table4=pd.DataFrame(purchase_df.groupby(['Gender'])[['Price']].sum())
table4['Total Purchase Value']=table4

#getting unique players for each gender category using groupby
table4['unique SN count']=purchase_df.groupby('Gender')['SN'].nunique()
table4
```

```python
table4['Average Purchase Price']=round(table4['Total Purchase Value']/purchase_count,2)
table4
```

```python
#Purchasing analysis(gender)
table4['Purchase Count']=purchase_df.groupby(['Gender'])[['SN']].count()
table4
```

```python

table4['Avg Total Purchase per Person']=table4['Total Purchase Value']/table4['unique SN count']
table4['Total Purchase Value'] = table4['Total Purchase Value'].map("${:.2f}".format)
table4['Avg Total Purchase per Person']=table4['Avg Total Purchase per Person'].map("${:.2f}".format)
table4['Average Purchase Price']=table4['Average Purchase Price'].map("${:.2f}".format)
table4
```

```python
#formatting- retaining the required columns
table4_final=['Purchase Count', 'Average Purchase Price', 'Total Purchase Value', 'Avg Total Purchase per Person']
```

```python
# Most Popular Items
#Get Purchase count for purchased items
popular_items_df = purchase_df.groupby(['Item ID','Item Name','Price']).size().reset_index(name = 'Purchase Count')
popular_items_df.rename(columns = {'Price':'Item Price'}, inplace = True)
```

```python
popular_items_df.head(5)
```

```python
popular_items_df['Total Purchase Value'] = popular_items_df['Item Price']*popular_items_df['Purchase Count']
popular_items_table = popular_items_df.sort_values(by=['Purchase Count'], ascending=False)
```

```python
popular_items_table.head(5)
```

```python
popular_items_table2 = popular_items_table[['Item ID','Item Name','Purchase Count','Item Price','Total Purchase Value']]
```

```python
popular_items_table2.head(5)
```
